```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set display options for pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)

# Load the datasets
food_security_data = pd.read_excel('processed_normalized_dataset.xlsx')
health_data = pd.read_excel('Cleaned_Health-Related_Indicators.xlsx')
selected_country_data = pd.read_excel('filtered_data_selected_countries.xlsx')

# Display the first few rows of each dataset
print("Food Security Data Sample:")
print(selected_country_data.head())
```

```python
print("\nHealth Data Sample:")
print(health_data.head())
```

```python
# Descriptive statistics for food security data
print("\nDescriptive Statistics for Food Security Data:")
print(food_security_data.describe())
```

```python
# Descriptive statistics for health data
print("\nDescriptive Statistics for Health Data:")
print(health_data.describe())
```

```python
# Visualizing individual variables - Histograms
selected_country_data.hist(figsize=(24, 18), bins=20)
plt.suptitle('Histograms of Food Security Indicators')
plt.show()
```

```python
# Visualizing individual variables - Box Plots
plt.figure(figsize=(6, 4))
sns.boxplot(data=food_security_data)
plt.title('Box Plot of Food Security Indicators')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze relationships - Scatter plot example (change columns as needed)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=health_data, x='sdg3_matmort', y='sdg3_u5mort')
plt.title('Maternal Mortality vs. Under-Five Mortality Rate')
plt.xlabel('Maternal Mortality (per 100,000 live births)')
plt.ylabel('Under-Five Mortality Rate (per 1,000 live births)')
plt.show()
```

```python
correlation_matrix = health_data[['sdg3_matmort', 'sdg3_neonat', 'sdg3_u5mort', 'sdg3_tb', 'sdg3_hiv', 'sdg3_ncds', 'sdg3_lifee', 'sdg3_fertility']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of Health Indicators')
plt.show()
```

```python
# Compare different countries - Bar chart
country_comparison = health_data.groupby('Country')['sdg3_lifee'].mean().sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=country_comparison.index, y=country_comparison.values)
plt.title('Average Life Expectancy by Country')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy (years)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Compare Maternal Mortality Rates by Country
maternal_mortality_comparison = health_data.groupby('Country')['sdg3_matmort'].mean().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=maternal_mortality_comparison.index, y=maternal_mortality_comparison.values, palette='plasma')
plt.title('Average Maternal Mortality by Country')
plt.xlabel('Country')
plt.ylabel('Average Maternal Mortality (per 100,000 live births)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot for Life Expectancy vs. Maternal Mortality
plt.figure(figsize=(10, 6))
sns.scatterplot(data=health_data, x='sdg3_matmort', y='sdg3_lifee', hue='Country', palette='deep')
plt.title('Life Expectancy vs. Maternal Mortality')
plt.xlabel('Maternal Mortality (per 100,000 live births)')
plt.ylabel('Life Expectancy (years)')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

```python
numeric_columns = health_data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()

#heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Health Indicators')
plt.show()
```
